### 1. Introduction to the Domain-specific area

Spam refers to unsolicited email messages that are sent in bulk. They are so prominent that by 2014, spam messages have grown to account for 90% of total email traffic, [1] with 100 billion spam messages being sent each day. There are many disadvantages caused by spam messages, all of which are born by the target recipient, making them a challenge for policy makers, internet users, and businesses. The following are some of the negative impacts of spam.

Firstly, spam is expensive because it causes communications overload. High volumes of spam consume valuable network resources as it blocks communication channels and creates traffic that has to be managed or funded by end users and Internet Service Providers (ISPs). For example, according the Alexander Ivanov, the President of the Russian Association of Networks and Services, spam caused a total loss of $55 million in traffic expenses to internet operators three years ago. [3] Hence, spam is a liability due to its high cost to its victim and minimal cost to the perpetrators.
Secondly, spam is a social threat as it reduces productivity. They are annoying as form of attention theft, as perpetrators seek to distract targets with their marketing content, even though they did not consent to view them and are given nothing in return. Other than its effect on a person’s emotional wellbeing, it causes clutter to a person’s mail or SMS system, and consumes valuable time to be sifted through or removed. Therefore spam reduces the productivity of a society. 
Thirdly, and arguably the most harmful negative impact comes from the criminalization of spam. This is commonly done through phishing software, spyware, and even ransomware. 

To start off, spam may contain links to phishing websites. Scammers pose as a legitimate source in an attempt to gain sensitive information such as usernames, passwords, or even credit card details. 
Nowadays, cybercriminals have become very creative in creating fake messages or websites that is completely undetectable, causing many to fall for their schemes.

Additionally, spyware refers to a software that a third party uses to track activity and gain information from the victim’s computer. It can be installed by unsuspecting victims simply clicking by on a malicious link. It can then track usernames, passwords, emails and any sensitive information which cybercriminals can then use to gather information about their bank details, allowing them to steal money. 

Lastly, ransomware refers to locking a computer or IT network and holding it at ransom. Once installed, it can lock files, making the computer unusable. It can be installed by clicking or downloading and unsuspecting file. A ransom note will then appear to request a ransom, usually in bitcoins, in exchange for autonomy over the computer back. [4] Therefore, this is bad as victims can be scammed out of a hefty sum of money. 

In conclusion, spam is a challenge as it only causes negative impacts to society. However, spammers are highly inventive and relentless, constantly designing more attractive bait to entice users into their traps. As a result, we can look towards text classification as a method to automatically detect spam messages and reduce its impacts on society.

### 4. Evaluation Methodology

To evaluate the outputs of my investigations, I will be using two evaluation metrics, namely accuracy and AUC-ROC. 

Both metrics use the concept of the confusion matrix for its calculations. It is a table layout that allows visualization and it describes the complete performance of the model. It involves the following 4 terms, which will be used differently in the algorithm for accuracy and AUC-ROC. 

1)	True positives (TP): Cases where actual and predicted values are positive
2)	True negatives (TN): Cases where the actual and predicted values are negative
3)	False positives (FP): Cases where the actual value is negative, but predicted values are positive
4)	False negatives (FN): Cases where the actual value is positive, but predicted values are negative

 
Figure 1 [7]

The confusion matrix table is illustrated in the image above.

Firstly, accuracy as an evaluation metric measures the total number of predictions, both positive and negative. It is one of the simplest form of evaluation. Accuracy can be calculated by using the values needed from the confusion matrix and implementing them in the formula shown below. 1 corresponds to the highest accuracy and 0 to the lowest. 


Figure 2 [8]

It can be applied by using the metrics module from scikit-learn library to implement the accuracy_score function, which will then use the formula to calculate accuracy. Hence, it uses metrics.accuracy_score() with the parameters y_test and y_prediction to calculate accuracy between the actual test and predicted results trained by the model.
 
Secondly, AUC-ROC refers to the area under curve (AUC) of the receive operator characteristic (ROC) curve. This is one of the most well-known metric for evaluation. ROC is a probability curve that visualises the trade-off between true positive rate (TPR) and false positive rate (FPR), to separate the signal from the noise. 
The TPR refers to the proportion of positive data that are correctly considered as positive, with regards to all positive data points. On the other hand, FPR refers to the proportion of negative data that were wrongly considered as positive, with regards to all negative data points. TPR and FPR are then computed at various threshold values to draw a graph, for both their ranges of 0 to 1. 

AUC is then considered as the area of the graph that is under the plotted line. It is the measure of the ability of a classifier to differentiate between classes. [7] AUC has a range of 0 to 1, and a higher AUC corresponds to a better ability of the model at differentiating between negative and positive classes. Visually, a curve that leans more to the y-axis performs better. The graph below shows when 0.5<AUC<1, which means that it can identify more true positives and negatives than false positives and negatives. Thus, classifier is performing relatively well at determining positive from negative class values. 


Figure 3 [7]

This evaluation metric can be applied by using the metrics module from scikit-learn library to implement the roc_auc_score function, to calculate AUC-ROC. Hence, it uses metrics.roc_auc_score() with the parameters y_test and y_prediction to calculate AUR-ROC between the actual test and predicted results trained by the model. 

Hence, I will be using accuracy and AUC-ROC to evaluate the output of my investigation.

### 5. Pre-processing

Preprocessing data refers to transforming raw data into well formed data, for mining analytics to be applied. This plays a significant role in increasing the success of the project, because raw data is often inconsistent in its formatting, or even incomplete.

### 5.1 Cleaning data

Since the class category is represented in string, it has to be mappped to integer values before it can be used in an algorithm for modeling. Thus, the code below shows a unique integer being assigned to each class category in a technique known as label encoding, where "ham" = 0, and "spam" = 1.

```python
# dropping columns that are null
message.dropna(how="any", inplace=True, axis=1)

# renaming column names
message.columns = ['Category', 'Message']

# convert label to integer variable
message['Category_num'] = message.Category.map({'ham':0, 'spam':1})

message.head()
```

### 5.2 Remove punctuation and stopwords

A preprocessing_text method that removes punctuations, stop words, and non alphabets is implemented. Stop words refers to a set of commonly used words in a language and are filtered out before preprocessing. It helps us narrow down the focus onto words that add more valuable insight to whether a message is considered spam.

```python
# importing and downloading the library nltk

# import nltk
# nltk.download ()

import string
from nltk.corpus import stopwords


def preprocessing_text(mess):
    
    """
    Function takes in string of text as the input, and 
        1. Removes all punctuation
        2. Removes all stopwords 
        3. Removes all non-alphabetic characters
    It returns clean text in a list form as the output
    """
    
    # removes punctuation
    nopunc = [char for char in mess if char not in string.punctuation]
    
    # remform string by join characters again
    nopunc = ''.join(nopunc)
    
    # adding a list of words to be included as stopwords
    stop_words = stopwords.words('english') + ['u', 'ü', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure']

    # remove stopwords and non-alphabetic characters
    return ' '.join([word for word in nopunc.split() if word.lower() not in stop_words and word.isalpha])
```

### 5.2 Tokenization

Tokenization refers to the process of changing normal text strings into a list of tokens, which are words that are provide more context to whether the message is spam or not.

```python
# apply the function to the dataset
message['Clean_message'] = message.Message.apply(preprocessing_text)
message.head()
```

#### Exploratory data analysis (EDA)
Finding the top 50 words used in spam and ham to help inform our model

```python
from collections import Counter

# changing each word into lowercase and splitting messages into words
wording = message[message.Category=='ham'].Clean_message.apply(lambda x: [word.lower() for word in x.split()])
hamword = Counter()


# update each message with counter
for messages in wording:
    hamword.update(message)
    
print(hamword.most_common(50))
```

### Splitting up data

```python
# defining X and y variables 
X = message.Clean_message
y = message.Category_num

# split X and y into training and testing sets of 80% to 20%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

print("Number of messages in X train:",X_train.shape)
print("Number of messages in X test:",X_test.shape)
print("Number of messages in y train:",y_train.shape)
print("Number of messages in y test:", y_test.shape)
```

### 5.3 Text representation: Bag of words

Bag of words refer to a statistical language model that uses word count to analyse text and documents. Text is represented as a bag of its own words and does not take into account word order or grammer, but keeping its multiplicity. It is simple to use and very efficient.

It works by building a sparse matrix that maps the occurrence of each word to the corpus vocabulary. Hence, the matrix that is mostly comprised of zeros. This format allows text to be converted into an interpretable encoding of linguistic information that the models can then implement. 

```python
# importing vectorizer from library
from sklearn.feature_extraction.text import CountVectorizer

# initialising the count vectorizer
vect = CountVectorizer()
vect.fit(X_train)

# creating a document-term matrix from learning training data vocabularly
X_train_dtm_val = vect.transform(X_train)

# fit and transform data
X_train_dtm_val = vect.fit_transform(X_train)

# creating a document-term matrix from testing data by using fitted vocabularly
X_test_dtm_val = vect.transform(X_test) 
```

### 5.4 Text representation: Tfidf

TF-IDF stands for term frequency-inverse document frequency, and bag of words can be improved by using tfidf as a text representation tool. The code below will run the vectors from bag of words through tfidf. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus. This is commonly used by search engines to rank the relevance of documents base on query input.

```python
# importing tfidf from library
from sklearn.feature_extraction.text import TfidfTransformer

# initialising the transformer
tfidf_transformer = TfidfTransformer()

# transforming bag of words into tfidf corpus
tfidf_transformer.fit(X_train_dtm_val)
tfidf_transformer.transform(X_train_dtm_val) 
```

### 6.2 Multinomial naive bayes

The multinomial naive bayes is good at handling discrete values as it describes the probability of observing frequency in a number of categories. It models the data distribuiton with a best-fit multinomial distribution and is good for text classification. Hence, it can be used as a baseline in spam detection, as it involves word counts and classifies the frequences of words within a document, making it suitable to be used as a benchmark to compare with other models. 

#### Using bag of words as the text classifier

```python
# import multinomial naive bayes model from library
from sklearn.naive_bayes import MultinomialNB

# initialise the model
mnb = MultinomialNB()

# train the model using training data
mnb.fit(X_train_dtm_val, y_train)

# make class predictions with testing data
y_predict_label = mnb.predict(X_test_dtm_val)

# find accuracy of label predictions
from sklearn import metrics
score = metrics.accuracy_score(y_test, y_predict_label)

print("The accuracy score for multinomial naive bayes model when using bag of words is", score)
```

```python
# print the confusion matrix

print("Confusion matrix for multinomial naive bayes model:")
metrics.confusion_matrix(y_test, y_predict_label)
```

```python
# find predicted probabilities for test values
y_prediction_prob = mnb.predict_proba(X_test_dtm_val)[:, 1]

# find AUC ROC value
score = metrics.roc_auc_score(y_test, y_prediction_prob)

print("The AUC ROC score for multinomial naive bayes model when using bag of words is", score)
```

#### Using tfidf as the text classifier

```python
# importing tfidf and pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline

# creating a pipleline to run input through count vectorizer, tfidf, an multinomial naive bayes
pipe_line = Pipeline([('bow', CountVectorizer()), 
                 ('tfid', TfidfTransformer()),  
                 ('model', MultinomialNB())]) # beautify way of fit transform

# fitting model with training data and using pipeline
pipe_line.fit(X_train, y_train)

# using the fitted model to predict with testing data
y_prediction = pipe_line.predict(X_test)

# find accuracy of predicted values
score = metrics.accuracy_score(y_test, y_prediction)
print("The accuracy score for multinomial naive bayes model when using tfidf is", score)
```

```python
# print the confusion matrix

print("Confusion matrix for multinomial naive bayes model:")
metrics.confusion_matrix(y_test, y_prediction)
```

### 7. Classification Approach

### 7.1 Logistic regression

Logistic regression as a model refers to using statistical analysis to predict binary classes based on a dataset. It is the appropriate regression analysis to conduct for dependent variables that are binary. It is used to describe the relationship between one or more independent variables and a dependent data variable.

```python
# import logistic regression from library
from sklearn.linear_model import LogisticRegression

# initialise the model
log_regression = LogisticRegression(solver='liblinear')

# train the model using training data
log_regression.fit(X_train_dtm_val, y_train)

# make class predictions with testing data
y_predict_label = log_regression.predict(X_test_dtm_val)

# find accuracy of label predictions
score = metrics.accuracy_score(y_test, y_predict_label)

print("The accuracy score for logistic regression model when using bag of words is", score)
```

```python
# print the confusion matrix

print("Confusion matrix for logistic regression model:")
metrics.confusion_matrix(y_test, y_predict_label)
```

```python
# find predicted probabilities for test values
y_prediction_prob = log_regression.predict_proba(X_test_dtm_val)[:, 1]

# find AUC ROC value
score = metrics.roc_auc_score(y_test, y_prediction_prob)

print("The AUC ROC score for logistic regression model when using bag of words is", score)
```

```python
# creating a pipleline to run input through count vectorizer, tfidf, and logistic regression
pipe_line = Pipeline([('bow', CountVectorizer()), 
                 ('tfid', TfidfTransformer()),  
                 ('model', LogisticRegression())])

# fitting model with training data and using pipeline
pipe_line.fit(X_train, y_train) 

# using the fitted model to predict with testing data
y_prediction = pipe_line.predict(X_test) #tfidf

# find accuracy of predicted values
score = metrics.accuracy_score(y_test, y_prediction)
print("The accuracy score for logistic regression model when using tfidf is", score)
```

### 7.2 Support Vector Machine
What Support vector machines do, is to not only draw a line between two classes here, but consider a region about the line of some given width. Here’s an example of what it can look like: 

```python
# import logistic regression from library
from sklearn.svm import SVC

# initialise the model
svc = SVC(kernel='linear',probability=True)

# train the model using training data
svc.fit(X_train_dtm_val, y_train)

# make class predictions with testing data
y_predict_label = svc.predict(X_test_dtm_val)

# find accuracy of label predictions
score = metrics.accuracy_score(y_test, y_predict_label)

print("The accuracy score for support vector machine model when using bag of words is", score)
```

```python
# print the confusion matrix

print("Confusion matrix for support vector machine model:")
metrics.confusion_matrix(y_test, y_predict_label)
```

```python
# find predicted probabilities for test values
y_prediction_prob = svc.predict_proba(X_test_dtm_val)[:, 1]

# find AUC ROC value
score = metrics.roc_auc_score(y_test, y_prediction_prob)

print("The AUC ROC score for support vector machine model when using bag of words is", score)
```

```python
# creating a pipleline to run input through count vectorizer, tfidf, an multinomial naive bayes
pipe_line = Pipeline([('bow', CountVectorizer()), 
                 ('tfid', TfidfTransformer()),  
                 ('model', SVC())])

# fitting model with training data and using pipeline
pipe_line.fit(X_train, y_train)

# using the fitted model to predict with testing data
y_prediction = pipe_line.predict(X_test)

# find accuracy of predicted values
score = metrics.accuracy_score(y_test, y_prediction)
print("The accuracy score for support vector machine model when using tfidf is", score)
```
